```python
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
df_standard23 = pd.read_csv('players_stat/standard(22-23).csv')
df_passing23 = pd.read_csv('players_stat/passing(22-23).csv')
df_creation23 = pd.read_csv('players_stat/creation(22-23).csv')
df_shooting23 = pd.read_csv('players_stat/shooting(22-23).csv')
df_defensive23 = pd.read_csv('players_stat/defensive(22-23).csv')
df_possession23 = pd.read_csv('players_stat/possession(22-23).csv')
df_keeper23 = pd.read_csv('players_stat/keeper(22-23).csv')
df_keeperadv23 = pd.read_csv('players_stat/keeperadv(22-23).csv')
df_playingtime23 = pd.read_csv('players_stat/playingtime(22-23).csv')
df_misc23 = pd.read_csv('players_stat/misc(22-23).csv')
```

```python
# add all the dataframes to a list
df_list = [df_standard23, df_passing23, df_creation23, df_shooting23, df_defensive23,
           df_possession23, df_keeper23, df_keeperadv23, df_playingtime23, df_misc23]
#loop to concat all df
for df in df_list:
    new_cols = [col for col in df.columns if col not in df_standard23.columns]
    df_standard23 = pd.concat([df_standard23, df[new_cols]], axis=1)
```

```python
# drop rows with NaN player names
df_standard23.dropna(subset=['Player'], inplace=True)
```

```python
# df_standard23 = df_standard23.rename(columns={"Goals":"G/90s"})
# df_21_22['Goals']= round(df_21_22['G/90s'] * df_21_22['90s'])
# df_21_22.Goals = df_21_22.Goals.astype(int)
```

```python
df_standard23 = df_standard23.rename(columns={"Player": "Name", "Squad": "Team"})
#Passing
df_standard23 = df_standard23.rename(columns={"Cmp": "TotCmp", "Att": "TotAtt", "Cmp%": "TotCmp%", })
df_standard23 = df_standard23.rename(columns={"Cmp.1": "Cmp.Short", "Cmp.2": "Cmp.Medium", "Cmp.3": "Cmp.Long", })
df_standard23 = df_standard23.rename(columns={"Att.1": "Att.Short", "Att.2": "Att.Medium", "Att.3": "Att.Long", })
df_standard23 = df_standard23.rename(columns={"Cmp%.1": "Cmp%.Short", "Cmp%.2": "Cmp%.Medium", "Cmp%.3": "Cmp%.Long",})
df_standard23 = df_standard23.rename(columns={"Gls.1": "Gls/90s", "Att": "TotAtt", "Cmp%": "TotCmp%", })
```

```python
# Standard stat
df_standard23 = df_standard23.rename(columns={"Gls.1": "Gls.Per90min",
                                              "Ast.1": "Ast.Per90min",
                                              "G+A.1": "G+A.Per90min",
                                              "G-PK.1": "G-PK.Per90min",
                                              "xG.1": "xG.Per90min",
                                              "xAG.1": "xAG.Per90min",
                                              "npxG.1": "npxG.Per90min",
                                              "npxG+xAG.1": "G-npxG+xAG.Per90min"})
# Goalkeeping & Advance Goalkeeping
#PK = Penalty kick
#GK = Goal Kick
df_standard23 = df_standard23.rename(columns={"AvgLen.1": "AvgLen.GK",
                                              "Save%.1": "Save%.PK",
                                              "Launch%.1": "Launch%.GK"})
# Goal and Shot Creation
#GCA = Goal creating actions
df_standard23 = df_standard23.rename(columns={"PassLive.1": "PassLive.GCA",
                                              "PassDead.1": "PassDead.GCA",
                                              "TO.1": "TO.GCA",
                                              "Sh.1": "Sh.GCA",
                                              "Fld.1": "Fld.GCA",
                                              "Def.1": "Def.GCA"})
# Defensive Action
#CHLG = challenge
df_standard23 = df_standard23.rename(columns={"Tkl.1": "Tkl.CHLG"})
# Playing time
df_standard23 = df_standard23.rename(columns={"On-Off.1": "On-Off.xG"})
```

```python
#do only once or RESTART AND RUN ALL
df_standard23.Nation = df_standard23.Nation.str.split(n=1).str.get(1)
df_standard23.Comp = df_standard23.Comp.str.split(n=1).str.get(1)
df_standard23.Age = df_standard23.Age.str.split('-').str.get(0)
df_standard23 = df_standard23.drop(['Matches', '-9999'], axis=1)
```

```python
temp = df_standard23[(df_standard23.Pos.str.startswith('FW') & df_standard23.Comp.str.contains('eng Premier League'))]
temp = temp[(temp.Starts > 5) & (temp.Min > 450) & (temp.Gls > 2)]
temp.shape
```

```python
temp['Gls-xG'] = temp['Gls'] - temp['xG']
```

```python
temp.sort_values(by='xG', ascending=False).head(10)[['Name','Gls','xG','Starts','Min']]
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='Gls', y='Gls-xG', data=temp.sort_values(by='xG', ascending=False).head(10), hue='Name')
for i, row in temp.sort_values(by='xG', ascending=False).head(10).iterrows():
    plt.text(row['Gls'], row['Gls-xG'], row['Name'], ha='center', va='center')
```
